```python
import os
import nbformat
from nbconvert import PythonExporter
from nbformat.reader import NotJSONError

# Base directory where the search for .ipynb files starts
# Replace this with the actual path to your base directory
# base_directory = '/path/to/your/base/directory'  # Replace with your directory
base_directory = '/Users/phil/Library/CloudStorage/Dropbox/GITHUB/notebook_poc/notebooks/'  # Example path


# New directory to store converted .py files
temp_directory = os.path.join(base_directory, 'temp_py_files')

# Create the temp_py_files directory if it doesn't exist
if not os.path.exists(temp_directory):
    os.makedirs(temp_directory)

# Function to convert a notebook to a Python script
def convert_notebook_to_script(notebook_path, output_path):
    try:
        # Load the notebook
        with open(notebook_path) as f:
            nb = nbformat.read(f, as_version=4)

        # Configure the Python exporter
        exporter = PythonExporter()

        # Convert to Python script
        body, _ = exporter.from_notebook_node(nb)

        # Save the Python script
        with open(output_path, 'w') as f:
            f.write(body)
    except NotJSONError as e:
        print(f"Error processing file {notebook_path}: {e}")
    except Exception as e:
        print(f"An error occurred with file {notebook_path}: {e}")

# Walk through the directory structure
for dirpath, dirnames, filenames in os.walk(base_directory):
    for filename in filenames:
        # Skip Jupyter's checkpoint files and ensure it's a valid notebook
        if filename.endswith('.ipynb') and not filename.endswith('-checkpoint.ipynb'):
            # Full path to the original .ipynb file
            notebook_path = os.path.join(dirpath, filename)

            # Convert .ipynb to .py
            converted_filename = filename.replace('.ipynb', '.py')
            converted_filepath = os.path.join(temp_directory, converted_filename)

            # Convert and save the notebook as a Python script
            convert_notebook_to_script(notebook_path, converted_filepath)

print("Conversion complete. Python files saved in:", temp_directory)
```
